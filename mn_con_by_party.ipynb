```python
import glob 
import pandas as pd
import matplotlib as mp
```

import and concatenate individual year/party contribution files

```python
path =r'C:\Users\m060040\Documents\pycode\mn_parties' # use your path
allFiles = glob.glob(path + "/*.csv")
mn_party_contrs = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_, skiprows=7, index_col=False, thousands=',')
    list_.append(df)
mn_party_contrs = pd.concat(list_)
```

```python
for mn_party_contrs['Donor Name'] in mn_party_contrs: 
    if mn_party_contrs[mn_party_contrs["Donor Name"].str.contains("DLF").bool()]:
        mn_party_contrs['Party'] = 'DFL'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('HRCC')].any():
        mn_party_contrs['Party'] = 'REP'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('RPM')].any():
        mn_party_contrs['Party'] = 'REP'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('Republican Party of Minn')].any():
        mn_party_contrs['Party'] = 'REP'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('Senate Victory Fund')].any():
        mn_party_contrs['Party'] = 'REP'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('Independence Party of Minn')].any():
        mn_party_contrs['Party'] = 'IPMN'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('IPMN')].any():
        mn_party_contrs['Party'] = 'IPMN'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('Green Party of Minn')].any():
        mn_party_contrs['Party'] = 'GPM'
    elif mn_party_contrs[mn_party_contrs['Donor Name'].str.contains('GPM')].any():
        mn_party_contrs['Party'] = 'GPM'
    else: mn_party_contrs['Party'] = 'UNK'
    
```

```python
mn_party_contrs['donate_date'] = pd.to_datetime(mn_party_contrs['Date'])
mn_party_contrs['donate_year']= mn_party_contrs['donate_date'].dt.year
mn_party_contrs.groupby(['donate_year','Recipient Committee']).sum()
```
